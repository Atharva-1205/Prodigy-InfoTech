# Importing Libraries

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading Data

```python
df = pd.read_csv('world_population.csv')
```

# Data Visualization

### Population of Countries by Continent

```python
plt.figure(figsize=(20, 8))
sns.barplot(x="Continent", y="2022 Population", data=df, estimator=sum, ci=None)
plt.title("Total Population by Continent (2022)")
plt.ylabel("Total Population (in billions)")
plt.xticks(rotation=0)
plt.show()
```

### Distribution of Population Densities

```python
plt.figure(figsize=(20, 8))
sns.histplot(df['Density (per km²)'], bins=30, kde=True)
plt.title("Distribution of Population Densities")
plt.xlabel("Density (per km²)")
plt.ylabel("Frequency")
plt.show()
```

### World Population Percentage by Continent

```python
plt.figure(figsize=(10, 10))
data_continent = df.groupby("Continent")["World Population Percentage"].sum()
plt.pie(data_continent, labels=data_continent.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title("World Population Percentage by Continent (2022)")
plt.show()
```

### Population Growth Over Time by Continent

```python
plt.figure(figsize=(20, 8))
data_melted = df.melt(id_vars=["Continent", "Country/Territory"], value_vars=["1970 Population", "1980 Population", "1990 Population", "2000 Population", "2010 Population",  "2015 Population", "2020 Population", "2022 Population"], var_name="Year", value_name="Population")
data_melted["Year"] = data_melted["Year"].str.extract('(\d+)').astype(int)
sns.lineplot(x="Year", y="Population", hue="Continent", estimator=sum, data=data_melted, ci=None)
plt.title("Population Growth Over Time by Continent")
plt.ylabel("Population (in billions)")
plt.xticks(rotation=0)
plt.show()
```

### Top 10 Most Populated Countries in 2022

```python
top_10_populous = df.nlargest(10, "2022 Population")
plt.figure(figsize=(20, 8))
sns.barplot(x="Country/Territory", y="2022 Population", data=top_10_populous)
plt.title("Top 10 Most Populous Countries in 2022")
plt.ylabel("Population (in billions)")
plt.xticks(rotation=0)
plt.show()
```

### Correlation Heatmap for Continuous Variables

```python
plt.figure(figsize=(12, 12))
continuous_vars = df[["2022 Population", "2020 Population", "2015 Population", "2010 Population", "2000 Population", "1990 Population", "Density (per km²)", "Growth Rate", "World Population Percentage"]]
corr = continuous_vars.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Continuous Variables")
plt.show()
```

### Population Density by Continent

```python
plt.figure(figsize=(20, 8))
sns.boxplot(x="Continent", y="Density (per km²)", data=df)
plt.title("Population Density by Continent")
plt.xticks(rotation=0)
plt.yscale("log")
plt.ylabel("Density (log scale)")
plt.show()
```

### Population according to Area

```python
plt.figure(figsize=(20, 8))
sns.scatterplot(x="Area (km²)", y="2022 Population", hue="Continent", size="Density (per km²)", sizes=(10, 200), data=df)
plt.title("Relationship between Population and Area (2022)")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Area (log scale)")
plt.ylabel("Population (log scale)")
plt.show()
```

### World Population from 1970 to 2022

```python
population_by_year = df[["1970 Population", "1980 Population", "1990 Population", "2000 Population", "2010 Population", "2015 Population", "2020 Population", "2022 Population"]].sum()
population_by_year_df = pd.DataFrame({
    'Year': [1970, 1980, 1990, 2000, 2010, 2015, 2020, 2022],
    'Population': population_by_year.values
})
plt.figure(figsize=(20, 8))
sns.barplot(x="Year", y="Population", data=population_by_year_df)
plt.title("Total World Population from 1970 to 2022")
plt.ylabel("Population (in billions)")
plt.xticks(rotation=0)
plt.show()
```
